## FLiBe

```python
T_list=np.arange(823, 1163, 30)
RP_LiF_FLiBe=[]
C_list_FLiBe=[]

x_LiF=0.66667
x_BeF2=1-x_LiF
xli=x_LiF/(3*x_BeF2+2*x_LiF)
xbe=x_BeF2/(3*x_BeF2+2*x_LiF)
xf=(x_LiF+2*x_BeF2)/(3*x_BeF2+2*x_LiF)

dbf=Database('Databases/MSTDB_V3-1.xml')
comps=['LI', 'F', 'BE']
points_dict={}
my_phases=['MSFL']
for phase_name in my_phases:
    mod = Model(dbf, comps, phase_name)
    points_dict[phase_name] = _sample_phase_constitution(mod, point_sample, True, 1000)

for t in T_list:
    lif_s=calculate(dbf, ['LI','F'], 'LIF_L1(LIQ)', P=101325, T=t)
    li_s=calculate(dbf, ['LI'], 'LI_S1(S)', P=101325, T=t)
    f2=calculate(dbf, ['F'], 'F2(G)', P=101325, T=t)
    deltaG0=lif_s.GM.values*2*2-f2.GM.values*2-2*li_s.GM.values
    conds = {v.N: 1, v.P: 101325, v.T: t, v.X('LI'): xli, v.X('BE'): xbe}
    eq = equilibrium(dbf, comps, ['MSFL'], conds, calc_opts={'points': points_dict})
    conds = {v.N: 1, v.P: 101325, v.T: t, v.X('LI'):0.5}
    refeq = equilibrium(dbf, ['LI', 'F'], ['MSFL'], conds)
    chempot_ref = refeq.MU.sel(component='LI').squeeze()+refeq.MU.sel(component='F').squeeze()
    chempot = (eq.MU.sel(component='LI').squeeze()+eq.MU.sel(component='F').squeeze())
    RP=(deltaG0[0][0][0][0]/2+chempot.values-chempot_ref.values)/96485.333 
    C_list_FLiBe.append(t-273.15)
    RP_LiF_FLiBe.append(RP)
#RP_LiF_FLiBe
```

### NiF2

```python
def convert_comp(x_NiF2):
    x_FLiNaK=1-x_NiF2
    x_LiF=x_FLiNaK*0.465
    x_NaF=x_FLiNaK*0.115
    x_KF=x_FLiNaK*0.42
    xni=x_NiF2/(x_NiF2*3+x_LiF*2+x_NaF*2+x_KF*2)
    xli=x_LiF/(x_NiF2*3+x_LiF*2+x_NaF*2+x_KF*2)
    xna=x_NaF/(x_NiF2*3+x_LiF*2+x_NaF*2+x_KF*2)
    xk=x_KF/(x_NiF2*3+x_LiF*2+x_NaF*2+x_KF*2)
    xf=(x_NiF2*2+x_LiF+x_NaF+x_KF)/(x_NiF2*3+x_LiF*2+x_NaF*2+x_KF*2)
    return xni, xli, xna, xk, xf
```

```python
T_list=np.arange(823, 1143, 20)
RP_NiF2_FLiNaK=[]
C_list=[]
dbf=Database('Databases/MSTDB_V3-1.xml')
xni, xli, xna, xk, xf = convert_comp(0.000001)
comps=['LI', 'F', 'NA', 'K', 'NI']
for t in T_list:
    nif2_s=calculate(dbf, ['NI','F'], 'NIF2_L1(LIQ)', P=101325, T=t)
    ni_s=calculate(dbf, ['NI'], 'NI_FCC(S)', P=101325, T=t)
    f2=calculate(dbf, ['F'], 'F2(G)', P=101325, T=t)
    deltaG0=nif2_s.GM.values*3-f2.GM.values*2-ni_s.GM.values
    conds = {v.N: 1, v.P: 101325, v.T: t, v.X('NI'): xni, v.X('LI'): xli, v.X('NA'): xna, v.X('K'): xk}
    eq = equilibrium(dbf, comps, ['MSFL'], conds)
    conds = {v.N: 1, v.P: 101325, v.T: t, v.X('NI'):1/3}
    refeq = equilibrium(dbf, ['NI', 'F'], ['MSFL'], conds)
    chempot_ref = refeq.MU.sel(component='NI').squeeze()+2*refeq.MU.sel(component='F').squeeze()
    chempot = eq.MU.sel(component='NI').squeeze()+2*eq.MU.sel(component='F').squeeze()
    RP=(deltaG0[0][0][0][0]+(chempot.values-chempot_ref.values))/96484/2 
    C_list.append(t-273.15)
    RP_NiF2_FLiNaK.append(RP)
```
